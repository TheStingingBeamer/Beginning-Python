```python
import pandas as pd
import numpy as np

dd=pd.read_csv("bigmart_data.csv")
dd=dd.dropna(how='any')
dd=dd.reset_index(drop=True)
dd.head()
```

# Aggregating Data using Pandas

```python
dd_by_item=dd.groupby('Item_Type')
```

```python
dd_by_item.first()
```

```python
dd_by_item.Item_Weight.mean()
```

```python
dd.groupby('Item_Type').Item_Weight.mean()
```

```python
pd.pivot_table(dd,index=['Item_Type','Item_Fat_Content'],values="Item_Weight", aggfunc=[np.mean,np.median])
```
