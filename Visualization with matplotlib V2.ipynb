```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

aa=pd.read_csv("bigmart_data.csv")
aa=aa.dropna(how='any')
aa[:5]
```

```python
price_by_item= aa.groupby('Item_Type').Item_MRP.mean()[:10]
price_by_item
```

```python
x=price_by_item.index.tolist()
y=price_by_item.values.tolist()

plt.title('Price by item type')
plt.xlabel('Item Type')
plt.ylabel('Average Price (In Rs.)')
plt.legend(labels=['Average Price'], loc='lower right')
plt.figure(figsize=(30,8))

plt.plot(x,y)
```

```python
plt.bar(x,y)
aa.columns
```
