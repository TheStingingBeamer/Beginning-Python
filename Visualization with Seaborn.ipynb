```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
aa=pd.read_csv('bigmart_data.csv')
aa=aa.dropna(how='any')
aa['Visibility_Scaled']=aa['Item_Visibility']*100
aa.head()
```

```python
sns.lineplot(data=aa[:50], x="Item_Weight", y="Item_MRP")
```

```python
sns.barplot(x="Item_Type", y="Item_MRP", data=aa[:5])
```

```python
sns.distplot(aa['Item_MRP'])
```

```python
sns.boxplot(aa['Item_Outlet_Sales'], orient='vertical')
```

```python
 sns.violinplot(aa['Item_Outlet_Sales'], orient='vertical', color='yellow')
```

```python
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", data=aa[:200])
```

```python
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", hue="Item_Type", data=aa[:200])
```

```python
sns.lineplot(data=aa[:100], x="Item_Weight", y="Item_MRP", hue="Outlet_Size")
```

```python
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", hue="Item_Type", kind='scatter', size="Visibility_Scaled", data=aa[:200])
```

# Category wise subplot

```python
sns.relplot(x="Item_MRP", y="Item_Outlet_Sales", col="Outlet_Size", hue='Item_Weight', kind='scatter', size="Visibility_Scaled", data=aa[:200])
```

# Advanced Categorical Plots in Seaborn

## Categorical scatterplots(stripplot (kind="strip") and swarnplot (kind="swarn") , distribution plots, estimate plots: all using catplot function  

```python
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales", kind='strip', data=aa[:250])
```

```python
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales", kind='swarm', data=aa[:250])
```

```python
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales", kind='box', data=aa)
```

```python
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales", kind='violin', data=aa)
```

```python
sns.catplot(x="Outlet_Size", y="Item_Outlet_Sales", kind='bar', data=aa)
```

# Density Plots : sns.kdeplot

```python
plt.figure(figsize=(7,7))
sns.kdeplot(aa['Item_Visibility'], shade=True)
```

```python
plt.figure(figsize=(7,7))
sns.kdeplot(aa['Item_MRP'], shade=True)
```

```python
plt.figure(figsize=(7,7))
sns.distplot(aa['Item_Outlet_Sales'])
```
